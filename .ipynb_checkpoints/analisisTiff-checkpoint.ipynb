```python
import openeo

# First, we connect to the back-end and authenticate. 
con = openeo.connect("openeo.dataspace.copernicus.eu")
con.authenticate_oidc()

# Now that we are connected, we can initialize our datacube object with the area of interest 
# and the time range of interest using Sentinel 1 data.
datacube1 = con.load_collection(
    "SENTINEL2_L2A",
    spatial_extent={"west": 5.14, "south": 51.17, "east": 5.17, "north": 51.19},
    temporal_extent = ["2023-02-01", "2023-04-30"],
    bands=["B01"],#["B02", "B04", "B08"],
    max_cloud_cover=35,
)

datacube1.download("RGB1.TIF")
datacube2 = con.load_collection(
    "SENTINEL2_L2A",
    spatial_extent={"west": 5.14, "south": 51.17, "east": 5.17, "north": 51.19},
    temporal_extent = ["2023-02-01", "2023-04-30"],
    bands=["B02"],#["B02", "B04", "B08"],
    max_cloud_cover=35,
)
datacube2.download("RGB2.TIF")
datacube3 = con.load_collection(
    "SENTINEL2_L2A",
    spatial_extent={"west": 5.14, "south": 51.17, "east": 5.17, "north": 51.19},
    temporal_extent = ["2023-02-01", "2023-04-30"],
    bands=["B03"],#["B02", "B04", "B08"],
    max_cloud_cover=35,
)
datacube3.download("RGB3.TIF")
datacube4 = con.load_collection(
    "SENTINEL2_L2A",
    spatial_extent={"west": 5.14, "south": 51.17, "east": 5.17, "north": 51.19},
    temporal_extent = ["2023-02-01", "2023-04-30"],
    bands=["B04"],#["B02", "B04", "B08"],
    max_cloud_cover=35,
)
datacube4.download("RGB4.TIF")
datacube5 = con.load_collection(
    "SENTINEL2_L2A",
    spatial_extent={"west": 5.14, "south": 51.17, "east": 5.17, "north": 51.19},
    temporal_extent = ["2023-02-01", "2023-04-30"],
    bands=["B05"],#["B02", "B04", "B08"],
    max_cloud_cover=35,
)
datacube5.download("RGB5.TIF")
datacube6 = con.load_collection(
    "SENTINEL2_L2A",
    spatial_extent={"west": 5.14, "south": 51.17, "east": 5.17, "north": 51.19},
    temporal_extent = ["2023-02-01", "2023-04-30"],
    bands=["B06"],#["B02", "B04", "B08"],
    max_cloud_cover=35,
)
datacube6.download("RGB6.TIF")
datacube7 = con.load_collection(
    "SENTINEL2_L2A",
    spatial_extent={"west": 5.14, "south": 51.17, "east": 5.17, "north": 51.19},
    temporal_extent = ["2023-02-01", "2023-04-30"],
    bands=["B07"],#["B02", "B04", "B08"],
    max_cloud_cover=35,
)
datacube7.download("RGB7.TIF")


```
